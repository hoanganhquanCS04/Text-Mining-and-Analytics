```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
df = pd.DataFrame(columns=['Name', 'Rate', 'Comment'])
URL = ["https://www.thegioididong.com/sac-cap#c=9518&m=37985,37986,24169,5301&o=7&pi=3",
       "https://www.thegioididong.com/sac-dtdd#c=57&m=37002,25047&o=7&pi=1",
       "https://www.thegioididong.com/tai-nghe-bluetooth#c=54&m=29875,2660,7710,24552,2391&p=56825&o=7&pi=2",
       "https://www.thegioididong.com/tai-nghe-co-day#c=54&m=29875,2391,2660,1842&p=56824&o=7&pi=0",
       "https://www.thegioididong.com/loa-laptop#c=2162&m=1176,14584,38245,14901,11925&o=7&pi=3",
       "https://www.thegioididong.com/camera-giam-sat-tp-link?g=camera-giam-sat",
       "https://www.thegioididong.com/dtdd#c=42&m=80,4832,2,1&o=7&pi=2"]
```

```python
all_links = []
```

```python
for url in URL:
  options = Options()
  options.add_argument("--headless")  # Xóa dòng này nếu muốn thấy trình duyệt
  driver = webdriver.Chrome(options=options)

  # Truy cập Wikipedia
  driver.get(url)
  soup = BeautifulSoup(driver.page_source, 'html')
  sanpham = soup.find_all('a', class_='main-contain')
  links = [link.get('href') for link in sanpham]
  links = ["https://www.thegioididong.com" + link for link in links]
  all_links.append(links)
  time.sleep(2)
```

```python
all_links
```

```python
all_links = [link for sublist in all_links for link in sublist]
```

```python
all_links
```

```python
len(all_links)
```

```python
all_rates = []
```

```python
for link in all_links: 
    driver.get(link)
    detailsoup = BeautifulSoup(driver.page_source, 'html.parser')

    rate_tag = detailsoup.find('a', class_='c-btn-rate')
    if rate_tag:  # chỉ append nếu tìm thấy
        all_rates.append("https://www.thegioididong.com" + rate_tag.get('href'))

    time.sleep(2)
```

```python
len(all_rates)
```

```python
pagination_links = []
```

```python
import pandas as pd
df = pd.DataFrame(columns=['Name', 'Rate', 'Comment'])
```

```python
df
```

```python
all_rates[1]
```

```python
for rate_link in all_rates:
    driver.get(rate_link)
    detail_rate = BeautifulSoup(driver.page_source, 'html.parser')
    pages = detail_rate.find_all('div', class_='pagcomment')
    list_name = [name.text for name in detail_rate.find_all("p", class_ = "cmt-top-name")]
    list_rate = [len(rate.find_all("i", class_="iconcmt-starbuy")) for rate in detail_rate.find_all('div', class_ = "cmt-top-star")]
    list_cmt = [cmt.text for cmt in detail_rate.find_all('p', class_ = 'cmt-txt')]
    new_df = pd.DataFrame({'Name': list_name, 'Rate': list_rate, 'Comment': list_cmt})
    df = pd.concat([df, new_df], ignore_index=True)
    if pages:  # chỉ xử lý nếu có thẻ 'div.pagcomment'
        page_links = pages[0].find_all("a")
        nums = [int(page.text) for page in page_links if page.text.isdigit()]

        if nums:  # chỉ xử lý nếu có số trang
            max_page = max(nums)
            for page in range(2, max_page + 1):
                page_url = rate_link + f"?page={page}"
                pagination_links.append(page_url)
                print(page_url)
    print(len(df))
    print(len(pagination_links))
    time.sleep(5)
```

```python
df_pagination = pd.DataFrame(columns=['Name', 'Rate', 'Comment'])
```

```python
index = 0
```

```python
for pagination_link in pagination_links:
  options = Options()
  options.add_argument("--headless")  # Xóa dòng này nếu muốn thấy trình duyệt
  driver = webdriver.Chrome(options=options)
  driver.get(pagination_link)
  time.sleep(6)
  detail_rate = BeautifulSoup(driver.page_source, 'html.parser')
  pages = detail_rate.find_all('div', class_='pagcomment')
  list_name = [name.text for name in detail_rate.find_all("p", class_ = "cmt-top-name")]
  list_rate = [len(rate.find_all("i", class_="iconcmt-starbuy")) for rate in detail_rate.find_all('div', class_ = "cmt-top-star")]
  list_cmt = [cmt.text for cmt in detail_rate.find_all('p', class_ = 'cmt-txt')]
  new_df = pd.DataFrame({'Name': list_name, 'Rate': list_rate, 'Comment': list_cmt})
  df_pagination = pd.concat([df_pagination, new_df], ignore_index=True)
  print(len(df_pagination))
  index += 1
              
```
